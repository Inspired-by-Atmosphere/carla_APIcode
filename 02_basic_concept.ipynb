创建摄像头并添加到车辆，分为以下四个步骤：
1. 创建摄像头蓝图并设置参数（分辨率、FOV、帧率）
2. 定义四个相对位姿（Transform）：前、后、左、右（或自定义角度）
3. 生成actor并attach_to自车
4. 注册回调函数通过listen()方法获取实时图像流

```python
# import numpy as np # 可再次导入（确保可用），用于图像数据处理
# import cv2  # 可再次导入OpenCV库，用于图像显示和处理

# 获取蓝图库对象
blueprint_library = world.get_blueprint_library()

# 存储摄像头对象和图像数据
# 初始化字典：存储摄像头Actor对象
cameras = {}
# 初始化字典：存储四个摄像头的图像数据，初始值为None
camera_images = {'front': None, 'rear': None, 'left': None, 'right': None}

## 定义摄像头配置参数（键值对形式，便于批量设置）
camera_config = {
    'image_size_x': '400', # 图像宽度（像素）
    'image_size_y': '300', #图像高度（像素）
    'fov': '110',  # 摄像头视场角
    'sensor_tick': '0.05'  # 20Hz采样率
}

# 配置四个摄像头位姿 (单位：米/度)
# 坐标系：X-向前，Y-向右，Z-向上
camera_positions = {
    'front': {'loc': (2.0, 0.0, 1.5),   'rot': (-15, 0, 0)},    # 前视摄像头：前2米，高1.5米，俯仰-15°（向下）
    'rear':  {'loc': (-2.0, 0.0, 1.5),  'rot': (-15, 180, 0)},  # 后视摄像头：后2米，高1.5米，偏航180°（向后）
    'left':  {'loc': (0.0, -1.0, 1.5),  'rot': (-15, -90, 0)},  # 左视摄像头：左1米，高1.5米，偏航-90°（向左）
    'right': {'loc': (0.0, 1.0, 1.5),   'rot': (-15, 90, 0)}    # 右视摄像头：右1米，高1.5米，偏航90°（向右）
}

def create_camera(world, bp_library, vehicle, location, rotation, name):
    """
    辅助函数：创建并配置摄像头传感器
    参数：
        world: 仿真世界对象
        bp_library: 蓝图库对象
        vehicle: 要附着的车辆Actor
        location: 摄像头相对车辆的位置元组 (x,y,z)
        rotation: 摄像头相对车辆的旋转元组 (pitch,yaw,roll)
        name: 摄像头名称（用于标识）
    返回：
        camera: 创建好的摄像头Actor对象
    """
    # 从蓝图库中查找RGB摄像头蓝图
    camera_bp = bp_library.find('sensor.camera.rgb')
    
    # 批量设置摄像头属性（遍历配置字典）
    for attr, value in camera_config.items():
        camera_bp.set_attribute(attr, value)
    
    #  # 定义摄像头相对于车辆的位姿变换对象 (x=前, y=右, z=上)
    transform = carla.Transform(
        carla.Location(x=location[0], y=location[1], z=location[2]),#位置
        carla.Rotation(pitch=rotation[0], yaw=rotation[1], roll=rotation[2])#旋转
    )
    
    # 生成摄像头并附加到车辆
    camera = world.spawn_actor(camera_bp, transform, attach_to=vehicle)
    
    # 定义图像处理的回调函数（摄像头产生数据时自动调用）
    def process_image(image, cam_name=name):
        
        """
        摄像头图像回调函数：将CARLA图像转换为OpenCV可用格式
        参数：
            image: CARLA图像对象
            cam_name: 摄像头名称（闭包传递）
        """
        # 将CARLA图像(raw_data)转换为numpy数组 (RGBA格式，4通道)
        array = np.frombuffer(image.raw_data, dtype=np.dtype("uint8"))
        #重塑数组形状为（高度，宽度，4）的三维数组
        array = np.reshape(array, (image.height, image.width, 4))  # RGBA
        
        # 去除Alpha通道（透明通道），保留前3个通道（RGB）
        array = array[:, :, :3]
        # OpenCV使用BGR格式，将RGB转换为BG
        array = array[:, :, ::-1]  
        # 将处理后的图像存储到全局字典中，按摄像头名称索引
        camera_images[cam_name] = array
        
        # 可选：保存到磁盘
        # image.save_to_disk(f'output/{cam_name}/{image.frame:06d}.png')
    
     # 注册回调函数：摄像头每帧数据都会调用process_image处理
    camera.listen(lambda image: process_image(image))
    # 返回创建好的摄像头对象
    return camera


# 遍历摄像头位置配置，批量创建四个摄像头
for name, cfg in camera_positions.items():
    # 调用辅助函数创建摄像头
    cam = create_camera(
        world, blueprint_library, ego_vehicle,
        cfg['loc'], cfg['rot'], name # 传入位置、旋转、名称
    )
    # 将摄像头对象存入字典，按名称索引
    cameras[name] = cam
    # 打印创建成功提示
    print(f"已创建摄像头: {name}")

# 主循环：实时显示四路摄像头视频流
try:
    # 无限循环，直到按下q键退出
    while True:
         # 同步仿真世界（确保数据更新）
        world.tick()  # 同步模式下需要手动tick
        
        # 检查所有摄像头图像是否都已获取（非None）
        if all(v is not None for v in camera_images.values()):
            # 拼接图像：上排（前+后），下排（左+右）
            top_row = np.hstack([camera_images['front'], camera_images['rear']]) # 水平拼接前、后视图
            bottom_row = np.hstack([camera_images['left'], camera_images['right']]) # 水平拼接左、右视图
            combined = np.vstack([top_row, bottom_row]) # 垂直拼接上下排
            
            # 显示拼接后的图像窗口
            cv2.imshow('Multi-Camera View (Front/Top, Rear/Bottom, Left/Right)', combined)
        # 检测键盘输入：按下q键（0xFF处理不同平台兼容性）退出循环    
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
finally:
    # 无论是否异常，最终执行资源清理
    print("销毁摄像头和车辆...")
    # 注释：按需启用销毁逻辑，避免重复执行
    # for cam in cameras.values():
    #     cam.stop()       # 停止摄像头数据采集
    #     cam.destroy()    # 销毁摄像头Actor
    # ego_vehicle.destroy()  # 销毁车辆Actor
    # cv2.destroyAllWindows()  # 关闭OpenCV窗口
```

与获取摄像头数据相似，IMU和GNSS的数据也是通过配置传感器参数、定义相对位置、选定附着车辆、注册回调函数来获取的。

```python
# 初始化列表：存储IMU和GNSS传感器Actor对象
sensors = []
# 初始化字典：存储传感器数据，初始值为None
sensor_data = {
    'imu': None,   # IMU数据
    'gnss': None   # GNSS数据
}
```

```python
def setup_imu(vehicle, world):
    """
    配置IMU（惯性测量单元）传感器
    参数：
        vehicle: 要附着的车辆Actor
        world: 仿真世界对象
    返回：
        imu: 创建好的IMU传感器Actor对象
    """
    # 从蓝图库中查找IMU传感器蓝图
    imu_bp = blueprint_library.find('sensor.other.imu')
    
    # 设置IMU属性：采样间隔0.02秒（50Hz）
    imu_bp.set_attribute('sensor_tick', '0.02')  # 50Hz 采样率
    # 可选：设置噪声参数（模拟真实传感器噪声）
    imu_bp.set_attribute('noise_accel_stddev_x', '0.1') # 加速度计X轴噪声标准差
    imu_bp.set_attribute('noise_gyro_stddev_x', '0.01') # 加速度计X轴噪声标准差
    
    # 定义IMU安装位置：车辆重心处（x=0,y=0,z=1.0米）
    imu_location = carla.Location(x=0.0, y=0.0, z=1.0)  # 车顶下方1米处
    imu_transform = carla.Transform(imu_location)
    
    # 生成IMU传感器Actor，并附着到车辆
    imu = world.spawn_actor(imu_bp, imu_transform, attach_to=vehicle)
    
    # 定义IMU数据回调函数
    def imu_callback(data):
        """
        IMU数据处理回调函数
        参数：
            data: CARLA IMU数据对象，包含加速度、陀螺仪、指南针数据
        """
        # 将IMU数据整理为字典，存储到全局sensor_data
        sensor_data['imu'] = {
            'frame': data.frame,         # 帧号
            'timestamp': data.timestamp, # 时间戳
            # 加速度数据（m/s²），包含重力加速度
            'accel': {
                'x': data.accelerometer.x,
                'y': data.accelerometer.y,
                'z': data.accelerometer.z
            },
            # 陀螺仪 (rad/s)
            'gyro': {
                'x': data.gyroscope.x,
                'y': data.gyroscope.y,
                'z': data.gyroscope.z
            },
            # 指南针 (度) - 相对于地理北极
            'compass': data.compass
        }
        
        # 实时打印（调试用）
        # print(f"IMU Accel: ({data.accelerometer.x:.2f}, "
        #       f"{data.accelerometer.y:.2f}, {data.accelerometer.z:.2f})")
    
    # 注册IMU回调函数
    imu.listen(imu_callback)
    # 返回IMU传感器对象
    return imu
```

```python
def setup_gnss(vehicle, world):
    """配置 GNSS 传感器"""
    gnss_bp = blueprint_library.find('sensor.other.gnss')
    
    # 设置属性
    gnss_bp.set_attribute('sensor_tick', '0.1')  # 10Hz（GNSS通常较低频）
    # 可选：设置GNSS噪声参数（模拟真实定位误差）
    gnss_bp.set_attribute('noise_alt_stddev', '0.1')    # 海拔噪声标准差（米）
    gnss_bp.set_attribute('noise_lat_stddev', '0.00001')# 纬度噪声标准差（度）
    gnss_bp.set_attribute('noise_lon_stddev', '0.00001')# 经度噪声标准差（度）
    
   # 定义GNSS安装位置：车顶（x=0,y=0,z=2.0米），信号更好
    gnss_location = carla.Location(x=0.0, y=0.0, z=2.0)  # 车顶上方
    gnss_transform = carla.Transform(gnss_location)
    
    # 生成GNSS传感器Actor，并附着到车辆
    gnss = world.spawn_actor(gnss_bp, gnss_transform, attach_to=vehicle)
    
    # 定义GNSS数据回调函数
    def gnss_callback(data):
        """
        GNSS数据处理回调函数
        参数：
            data: CARLA GNSS数据对象，包含经纬度、海拔
        """
        # 将GNSS数据整理为字典，存储到全局sensor_data
        sensor_data['gnss'] = {
            'frame': data.frame,            # 帧号
            'timestamp': data.timestamp,    # 时间戳
            'latitude': data.latitude,      # 纬度
            'longitude': data.longitude,    # 经度
            'altitude': data.altitude       # 海拔（米）
        }
        
        # 将GNSS经纬度转换为 CARLA 世界坐标（可选）
        # location = carla.Location(
        #     x=data.longitude,
        #     y=data.latitude,
        #     z=data.altitude
        # )
        # 实时打印GNSS数据（调试用）
        # print(f"GNSS: Lat {data.latitude:.6f}, Lon {data.longitude:.6f}, "
        #       f"Alt {data.altitude:.2f}m")
    
    # 注册GNSS回调函数
    gnss.listen(gnss_callback)
    # 返回GNSS传感器对象
    return gnss
```

将全局坐标的速度转换为车辆坐标的速度

```python
def get_vehicle_speeds(vehicle):
    """
    计算车辆本地坐标系下的速度分量
    车辆坐标系定义：x-向前，y-向右，z-向上
    参数：
        vehicle: 车辆Actor对象
    返回：
        (纵向速度, 横向速度, 垂直速度) - 单位 m/s
    """
    # 获取世界坐标系下的速度向量
    vel_world = vehicle.get_velocity()
    # 转换为numpy数组 [x,y,z]
    vel_vector = np.array([vel_world.x, vel_world.y, vel_world.z])
    
    # 获取车辆的偏航角（yaw），转换为弧度（math库用弧度计算）
    transform = vehicle.get_transform()
    yaw = math.radians(transform.rotation.yaw)
    
    # 构建旋转矩阵（世界坐标系 → 车辆坐标系）
    # 仅考虑偏航角（平面旋转），忽略俯仰和翻滚
    cos_yaw = math.cos(yaw)
    sin_yaw = math.sin(yaw)
    
    # 旋转矩阵（逆矩阵，世界→车辆）
    R = np.array([
        [cos_yaw, sin_yaw, 0],    # 车辆x轴（前）在世界坐标系的投影
        [-sin_yaw, cos_yaw, 0],   # 车辆y轴（右）在世界坐标系的投影
        [0, 0, 1]                 # 车辆z轴（上）与世界z轴一致
    ])
    
    # 矩阵乘法：世界速度向量 × 旋转矩阵 = 车辆本地速度向量
    vel_local = R @ vel_vector
    
    # 返回纵向（x）、横向（y）、垂直（z）速度
    return vel_local[0], vel_local[1], vel_local[2]
```

```python
# 调用函数计算车辆本地速度
v_long, v_lat, v_vert = get_vehicle_speeds(ego_vehicle)
# 打印本地速度信息
print(f"纵向速度: {v_long:.2f} m/s, 横向速度: {v_lat:.2f} m/s")
```

```python
import carla          # 再次导入（确保可用）
import numpy as np    # 再次导入（确保可用）
import open3d as o3d  # 导入Open3D库，用于点云可视化和保存
from datetime import datetime  # 导入时间模块，用于时间戳

class LidarManager:
    """
    激光雷达管理类：封装激光雷达的配置、数据处理、保存等功能
    """
    def __init__(self, world, vehicle):
        """
        构造函数：初始化激光雷达管理器
        参数：
            world: 仿真世界对象
            vehicle: 要附着的车辆Actor对象
        """
        self.world = world                  # 保存仿真世界对象
        self.vehicle = vehicle              # 保存车辆对象
        self.lidar = None                   # 激光雷达Actor对象（初始为空）
        self.point_cloud_data = None        # 存储最新点云数据（初始为空）
        
    def setup_lidar(self):
        """配置并生成激光雷达传感器"""
        # 获取蓝图库对象
        blueprint_library = self.world.get_blueprint_library()
        # 查找激光雷达蓝图（ray_cast类型）
        lidar_bp = blueprint_library.find('sensor.lidar.ray_cast')
        
        # 配置激光雷达关键参数
        lidar_bp.set_attribute('channels', '64')           # 激光雷达线束数（64线）
        lidar_bp.set_attribute('range', '100.0')           # 最大探测范围（100米）
        lidar_bp.set_attribute('points_per_second', '1000000')  # 每秒点云数量（100万点/秒）
        lidar_bp.set_attribute('rotation_frequency', '10.0')    # 旋转频率（10Hz，每秒10圈）
        lidar_bp.set_attribute('upper_fov', '15.0')        # 垂直视野上边界（+15°）
        lidar_bp.set_attribute('lower_fov', '-30.0')       # 垂直视野下边界（-30°）
        lidar_bp.set_attribute('horizontal_fov', '360.0')  # 水平视野（360°全向）
        
        # 可选：设置激光雷达噪声参数（模拟真实传感器）
        # lidar_bp.set_attribute('dropoff_general_rate', '0.0')  # 通用丢包率
        # lidar_bp.set_attribute('noise_stddev', '0.0')          # 距离测量噪声标准差
        
        # 定义激光雷达安装位置：车顶中心（x=0,y=0,z=2.4米）
        lidar_transform = carla.Transform(
            carla.Location(x=0.0, y=0.0, z=2.4),
            carla.Rotation(pitch=0, yaw=0, roll=0)
        )
        
        # 生成激光雷达Actor，并附着到车辆
        self.lidar = self.world.spawn_actor(
            lidar_bp, 
            lidar_transform, 
            attach_to=self.vehicle
        )
        
        # 注册点云数据回调函数
        self.lidar.listen(lambda data: self._process_lidar(data))
        # 打印初始化成功提示
        print("激光雷达已启动，频率 10Hz，64线")
        
    def _process_lidar(self, data):
        """
        私有方法：处理激光雷达点云数据
        参数：
            data: CARLA激光雷达数据对象
        """
        # 将原始字节数据转换为float32类型的numpy数组
        points = np.frombuffer(data.raw_data, dtype=np.float32)
        
        # 重塑数组为N×4的格式：每行对应一个点（x,y,z,intensity）
        points = np.reshape(points, (int(points.shape[0] / 4), 4))
        
        # 保存点云数据（包含元数据）
        self.point_cloud_data = {
            'frame': data.frame,                          # 帧号
            'timestamp': data.timestamp,                  # 时间戳
            'transform': data.transform,                  # 激光雷达的位姿
            'points': points,                             # 点云数据（N×4）
            'location': np.array([data.transform.location.x,
                                 data.transform.location.y,
                                 data.transform.location.z])  # 激光雷达位置
        }
        
        # 可选：实时打印点云数量（调试用）
        print(f"Frame {data.frame}: 接收到 {len(points)} 个点")
        
    def save_point_cloud(self, filename=None):
        """
        保存点云数据为PCD格式（Open3D支持）
        参数：
            filename: 保存文件名（可选，默认按帧号命名）
        """
        # 检查是否有点云数据
        if self.point_cloud_data is None:
            print("无点云数据")
            return
            
        # 自动生成文件名（如果未指定）
        if filename is None:
            filename = f"lidar_{self.point_cloud_data['frame']}.pcd"
            
        # 获取点云数据（x,y,z,intensity）
        points = self.point_cloud_data['points']
        
        # 创建Open3D点云对象
        pcd = o3d.geometry.PointCloud()
        # 设置点云坐标（仅取前3列：x,y,z）
        pcd.points = o3d.utility.Vector3dVector(points[:, :3])
        
        # 可选：将强度值转换为颜色（归一化到0-1）
        intensities = points[:, 3]                          # 获取强度列
        colors = np.tile(intensities[:, np.newaxis], (1, 3)) / 255.0  # 扩展为RGB三通道
        pcd.colors = o3d.utility.Vector3dVector(colors)     # 设置点云颜色
        
        # 保存点云为PCD文件
        o3d.io.write_point_cloud(filename, pcd)
        # 打印保存成功提示
        print(f"点云已保存: {filename}")
        
    def destroy(self):
        """销毁激光雷达Actor，释放资源"""
        if self.lidar:
            self.lidar.stop()          # 停止点云数据采集
            self.lidar.destroy()       # 销毁激光雷达Actor
            print("激光雷达已销毁")

# 使用示例：创建激光雷达管理器并运行
# 连接CARLA服务器（确保与之前的客户端一致）
client = carla.Client('localhost', 2000)
world = client.get_world()

# 生成车辆（假设已有ego_vehicle，实际使用时需确保车辆已创建）
# ego_vehicle = world.spawn_actor(...)

# 创建激光雷达管理器实例
lidar_mgr = LidarManager(world, ego_vehicle)
# 配置并启动激光雷达
lidar_mgr.setup_lidar()

try:
    # 主循环：保持运行，直到手动终止
    while True:
        # 同步仿真世界
        world.tick()
            
finally:
    # 资源清理：销毁激光雷达
    lidar_mgr.destroy()
```
